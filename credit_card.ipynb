```python
df = pd.read_csv('/content/drive/MyDrive/customer_data.csv')
df1=pd.read_csv('/content/drive/MyDrive/payment_data.csv')
merged_df = df.merge(df1, on='id')
```

```python
merged_df.drop(columns=['update_date','report_date'],inplace=True)
```

```python
merged_df.isnull().sum()
```

```python
from sklearn.impute import KNNImputer
impute = KNNImputer(n_neighbors= 3)
cols = ['prod_limit','highest_balance','fea_2']
merged_df[cols] = merged_df[cols].round()
for i in cols:
    merged_df[i] = impute.fit_transform(merged_df[[i]])
```

```python
corr_matrix = merged_df.corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
plt.figure(figsize=(20, 9))
plt.title("Correlation Heatmap", fontsize=16)
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr_matrix, mask=mask, cmap=cmap, annot=True, fmt=".2f", linewidths=0.5)
plt.xticks(rotation=45)
plt.show()
```

```python
original_id=merged_df['id']
```

```python
id_counts = {}
for id in original_id:
    if id in id_counts:
        id_counts[id] += 1
    else:
        id_counts[id] = 1
```

```python
new_ids = {}
for id, count in id_counts.items():
    if count > 1:
        new_ids[id] = [f"{id}_{i}" for i in range(1, count + 1)]
```

```python
updated_ids = []
for id in original_id:
    if id in new_ids:
        updated_ids.append(new_ids[id].pop(0))
    else:
        updated_ids.append(id)
```

```python
x = merged_df.drop(columns=['label'],axis=1)
y = merged_df['label']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y, test_size= 0.2)
```

```python
from sklearn.linear_model  import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
model1 = RandomForestClassifier()
model1.fit(X_train, y_train)
y_pred = model.predict(X_test)
predictions = model1.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
accuracy1 = accuracy_score(y_test,predictions)
print('logisticRegression:' )
print("Accuracy:", accuracy)
print()
print('RandomForestClassifier:' )
print("Accuracy:", accuracy1)
```

```python
new_data_predictions_logistic = model.predict(X_test)
for i in range(15):
    predicted_label = new_data_predictions_logistic[i]
    actual_label = y_test.iloc[i]
    print("Instance", i+1)
    print("Predicted Label:", predicted_label)
    print("Actual Label:", actual_label)
    print("✅" if predicted_label == actual_label else "❌")
    print()
```
